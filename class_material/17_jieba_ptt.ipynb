```python
import glob     #曲目錄下所有檔案名稱
import jieba

jieba.load_userdict("./mydict.txt")
with open("./stop_words.txt", "r") as f:
    stop_words_list = [w.replace("\n", "") for w in f.readlines()] 
    # 把所有停用字放到一list中，並刪除換行符號
```

```python
stop_words_list
```

```python
for file_path in glob.glob("./res_gossiping/*.txt"):
    print(file_path)
```

```python
all_str = "" # 製作一空字串

for file_path in glob.glob("./res_gossiping/*.txt"):
    with open(file_path, "r", encoding="utf-8") as f:
        article_str = f.read().split("---split---")[0]
        all_str += article_str    # 將文章都存在該字串中
```

```python
print(all_str)
```

```python
word_generator = jieba.cut(all_str)
```

```python
# word_list = list(word_generator) # 將generator轉為一個list
word_list = [w for w in word_generator if (len(w) > 1) and (w not in stop_words_list)]
# 必須大於1，並且不在停用字字點內
word_list
```

```python
# 計算詞出現的次數，製成一字典

word_count = {}   # 空字典

for word in word_list:
    if word not in word_count:
        word_count[word] = 1   # 讓字典的value等於1
    else:
        word_count[word] += 1

print(word_count)
```

```python
word_count
```

```python
# 進行排序，因字典無法排序，故先轉為list

word_count_list = [ (k, v) for k, v in word_count.items()]  # .items()同時取出字典的key+value
word_count_list
```

```python
word_count_list.sort(key=lambda x: x[1], reverse=True)  # x指的是每個tuple
```
